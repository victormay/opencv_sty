```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def show(img,T=1000,C=True):
    cv2.namedWindow("show",cv2.WINDOW_NORMAL)
    cv2.imshow("show",img)
    if cv2.waitKey(T):
        if C:
            cv2.destroyWindow("show")
```

```python
def line_demo(name):
    img = np.zeros((512,512,3),dtype=np.uint8)
    cv2.line(img,(0,0),(512,512),(255,0,0),2,cv2.LINE_AA)
    cv2.imwrite(name,img)
    show(img,0)
```

```python
line_demo("lineaa.jpg")
```

```python
def rec_demo():
    img = np.zeros((512,512,3),dtype=np.uint8)
    cv2.rectangle(img,(50,50),(500,500),(0,255,0),2)
    show(img,0)
rec_demo()
```

```python
def circle_demo():
    img = np.zeros((512,512,3),dtype=np.uint8)
    cv2.circle(img,(250,250),200,(0,255,0),4,cv2.LINE_AA)
    show(img,0)
circle_demo()
```

```python
def ellipse_demo():
    img = np.zeros((512,512,3),dtype=np.uint8)
    cv2.ellipse(img,(250,250),(100,50),0,0,360,(0,255,0),4,cv2.LINE_AA)
    show(img,0)
ellipse_demo()
```

```python
def lines_demo():
    img = np.zeros((512,512,3),dtype=np.uint8)
    pts = np.array([[10,5],[20,30],[70,20],[50,10]], np.int32) 
    pts = pts.reshape((-1,1,2)) 
    cv2.polylines(img,[pts],True,(0,255,0))
    show(img,0)
lines_demo()
```

```python
def draw_circle(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDBLCLK:
        cv2.circle(img,(x,y),100,(255,0,0),-1)
img = np.zeros((512,512,3), np.uint8)
cv2.namedWindow('image')
cv2.setMouseCallback('image',draw_circle)
while(1):
    cv2.imshow('image',img)
    if cv2.waitKey(20) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```

```python
def call_back_1():
    def demo(event,x,y,flags,param):
        if event==cv2.EVENT_LBUTTONDOWN:
            print(x,y)
    img = np.zeros((512,512,3), np.uint8)
    cv2.namedWindow('image')
    cv2.setMouseCallback('image',demo)
    while(1):
        cv2.imshow('image',img)
        if cv2.waitKey(20) & 0xFF == 27:
            break
    cv2.destroyAllWindows()
        
call_back_1()
```

```python
def call_back_2():
    
    x1,y1 = (0,0)
    drawing = False
    def draw_rec(event,x,y,flags,param):
        global x1,y1,drawing
        if event==cv2.EVENT_LBUTTONDOWN:
            x1,y1 = x,y
            drawing = True
        elif event==cv2.EVENT_MOUSEMOVE and flags==cv2.EVENT_FLAG_LBUTTON:
            if drawing:
                cv2.rectangle(img,(x1,y1),(x,y),(0,0,255),-1)
        elif event==cv2.EVENT_LBUTTONUP:
            drawing = False
    img = np.zeros((512,512,3), np.uint8)
    cv2.namedWindow('image')
    cv2.setMouseCallback('image',draw_rec)
    while True:
        cv2.imshow('image',img)
        if cv2.waitKey(20) & 0xFF == 27:
            break
    cv2.destroyAllWindows()
    
call_back_2()
```
