```python
import cv2
import numpy as np
```

```python
def show(img,T=1000,C=True):
    cv2.namedWindow("show",cv2.WINDOW_NORMAL)
    cv2.imshow("show",img)
    if cv2.waitKey(T) & 0xFF==27:
        if C:
            cv2.destroyWindow("show")
```

```python
img = cv2.imread("windows.jpg",1)
show(img,0)
```

```python
def trackbar_1():
    
    drawing = False
    color = (0,0,0)
    
    def draw_line(event,x,y,flags,param):
        global drawing
        if event==cv2.EVENT_LBUTTONDOWN:
            drawing = True
        elif event==cv2.EVENT_MOUSEMOVE and flags==cv2.EVENT_FLAG_LBUTTON:
            cv2.circle(img,(x,y),2,color,2)
            
    def r_(x):
        cv2.putText(img,)
    
    def g_(x):
        cv2.putText()
    
    def b_(x):
        cv2.putText()
    
    img  = np.zeros((512,512,3),dtype=np.uint8)
    cv2.namedWindow("trackbar_1",cv2.WINDOW_NORMAL)
    
    cv2.createTrackbar('R','trackbar_1',0,255,nothing)
    cv2.createTrackbar('G','trackbar_1',0,255,nothing)
    cv2.createTrackbar('B','trackbar_1',0,255,nothing)
    cv2.setMouseCallback("trackbar_1",draw_line,color)
    
    while True:
        cv2.imshow("trackbar_1",img)
        if cv2.waitKey(15) & 0xFF==27:
            break
        r = cv2.getTrackbarPos('R','trackbar_1')
        g = cv2.getTrackbarPos('G','trackbar_1')
        b = cv2.getTrackbarPos('B','trackbar_1')
        color = (b,g,r)
#         print(color)
    cv2.destroyWindow("trackbar_1")

trackbar_1()
```

```python
def trackbar_2():
    def nothing(x):
        pass
    img = np.zeros((640,480,3),np.uint8)
    cv2.namedWindow("trackbar_2",cv2.WINDOW_NORMAL)
    cv2.createTrackbar('R','trackbar_2',0,255,nothing)
    cv2.createTrackbar('G','trackbar_2',0,255,nothing)
    cv2.createTrackbar('B','trackbar_2',0,255,nothing)
    switch = '0 : OFF \n1 : ON'
    cv2.createTrackbar(switch, 'trackbar_2',0,1,nothing)
    while True:
        cv2.imshow("trackbar_2",img)
        if cv2.waitKey(15) & 0xFF==27:
            break
        r = cv2.getTrackbarPos('R','trackbar_2')
        g = cv2.getTrackbarPos('G','trackbar_2')
        b = cv2.getTrackbarPos('B','trackbar_2')
        s = cv2.getTrackbarPos(switch,'trackbar_2')
        if s == 0:
            img[:] = 0
        else:
            img[:] = [b,g,r]
    cv2.destroyWindow("trackbar_2")

trackbar_2()
```
